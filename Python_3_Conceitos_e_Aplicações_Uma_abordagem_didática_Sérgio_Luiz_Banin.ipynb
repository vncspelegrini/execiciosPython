pag 95 - Como foi visto, a execução desse programa para N = 8 gerou uma lista con-tendo duas ocorrências do valor 50. Nada foi dito no enunciado sobre isso. A tarefa é alterar esse programa de modo que não haja valores duplicados na lista gerada. Dica: lembre-se dos operadores in e not in.

```python
from random import randint
try:
  N = int(input('Digite a quantidade de números: '))
except:
  print('Você deve digitar um número inteiro')
Lista = []
Count = 0
while Count < N:
  Num = randint(10, 50)
  if Num not in Lista:
    Lista.append(Num)
    Count += 1
Lista.sort()
print(Lista)
```

pag 140 14. Faça uma pesquisa sobre o Algoritmo de Ordenação Quicksort. Implemente uma função recursiva que use esse algoritmo para organizar a lista L de forma crescente. Escreva um programa para testar a função.

pag 183 4. Repita o exercício anterior trocando o método de ordenação, para o método recursivo Quicksort (faça uma pesquisa sobre esse algoritmo). O Quicksort é um algoritmo consideravelmente mais veloz, na maioria dos casos, que o Bubble Sort. Use os programas 3 e 4 para constatar esse fato.


```python
def quicksort(arr):
    if len(arr) <= 1:
        return arr
    else:
        pivot = arr[0]
        less = [x for x in arr[1:] if x <= pivot]
        greater = [x for x in arr[1:] if x > pivot]
        return quicksort(less) + [pivot] + quicksort(greater)

lista = []
arq = open('numeros.txt' , 'r')
for s in arq:
  lista.append(s)
arq.close()

lista_ordenada = quicksort(lista)

arq = open('numeros.txt', 'w')
for i in lista_ordenada:
  arq.write(f'{str(i)}\n')
arq.close()

for s in open('numeros.txt'):
  print(s)
```

```python
import sqlite3

arq = open('cadastro.txt', 'w')

while True:
  codigo = input('Digite o código ou espaço para sair: ')
  qtde = input('Digite a quantidade ou espaço para sair: ')
  pccompra = input('Digite o preço de compra ou espaço para sair: ')
  pcvenda = input('Digite o preço de venda ou espaço para sair: ')
  arq.write(f'{codigo};{qtde};{pccompra};{pcvenda}\n')
  a = input('Deseja continuar? (s/n)')
  if a == 'n':
    break
arq.close()

conexao = sqlite3.connect('loja.db')
cursor = conexao.cursor()

sql = '''
  create table vendas (codigo integer primary key,
  qtde integer,
  pccompra real,
  pcvenda real
  )'''
cursor.execute(sql)
conexao.commit()

arq = open('cadastro.txt', 'r')
for i in arq:
  codigo, qtde, pccompra, pcvenda = i.split(';')
  sql = '''
  insert into vendas (codigo, qtde, pccompra, pcvenda)
  values (?, ?, ?, ?)
  '''
  cursor = conexao.cursor()
  cursor.execute(sql, (codigo, qtde, pccompra, pcvenda))
  conexao.commit()
arq.close()

```
